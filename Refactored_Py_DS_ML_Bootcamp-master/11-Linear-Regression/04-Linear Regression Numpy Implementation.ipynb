## Numpy Interpretation of the Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
def getSlope(x, y):
    ## sum[(xi -x)*(yi - y)] / sum[(xi - x)^2]
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    numerator = 0
    denominator = 0
    for i in range(len(x)):
        numerator += (x[i] - x_mean)*(y[i] - y_mean)
        denominator += (x[i] - x_mean)**2
    return numerator / denominator

def getIntercept(x, y):
    ## (sum(y)*sum(x^2) - sum(x)*sum(y)) / n*sum(x^2) - sum(x)^2
    x_sum = np.sum(x)
    x_squared_sum = np.sum(x**2)
    y_sum = np.sum(y)
    intercept = (y_sum*x_squared_sum - x_sum*np.sum(x*y)) / (len(x)*x_squared_sum - x_sum**2)
    return intercept
```

```python
x = np.array([1.5,2,2.5,3,3.5,4,4.5,5,5.5,6])
y = np.array([10.35,12.3,13,14.0,16,17,18.2,20,20.7,22.5])
slope = getSlope(x, y)
intercept = getIntercept(x, y)
print("Slope: ", slope)
print("Intercept: ", intercept)
```

```python
mn=np.min(x)
mx=np.max(x)
x1=np.linspace(mn,mx,500)
y1=slope*x1+intercept
plt.plot(x,y,'ob')
plt.plot(x1,y1,'-r')
plt.show()
```
